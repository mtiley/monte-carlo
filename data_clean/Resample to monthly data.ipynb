```python
import pandas as pd
import datetime
```

```python
mdata = pd.read_csv("interpolated_feedwheat.csv")
```

```python
mdata
```

```python
mdata["Date"] = pd.to_datetime(mdata["Date"].astype("str"), format="%d/%m/%Y")
```

```python
mdata
```

```python
monthly = mdata.resample("M", on="Date").mean()
```

```python
monthly
```

```python
monthly.to_csv("monthly_feedwheat_interpd.csv")
```
