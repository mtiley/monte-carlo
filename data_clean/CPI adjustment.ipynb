##  CPI deflation

```python
import pandas as pd
import datetime
```

```python
data1 = pd.read_csv("inner_join.csv")
```

```python
data1
```

```python
cpi = pd.read_csv("CPI.csv")
```

```python
cpi
```

```python
def generatedict(df):
    return pd.Series(df.CPIDec20.values,index=df.Quarter).to_dict()
```

```python
cpi_dict1 = generatedict(cpi)
```

```python
def get_CPI(date, cpi_dict = cpi_dict1):
    ## assume string e.g. "04/10/1948"
    month = date[-7:-5]
    
    if 1 <= int(month) <= 3: ## convert to correct quarter end date
        month = "03"
    elif 4 <= int(month) <= 6:
        month = "06"
    elif 7 <= int(month) <= 9:
        month = "09"
    elif 10 <= int(month) <= 12:
        month = "12"
    
    date = "1/" + month + "/" + date[-4:] 
    
    return cpi_dict[date]
    
    
```

```python
get_CPI("04/11/2009")
```

```python
newcol = []

for date in data1["Date"]:
    newcol.append(get_CPI(date))
```

```python
newcol
```

```python
data1["CPI (Dec20$)"] = newcol
```

```python
data1.to_csv("inner_join_with_CPI.csv")
```
