```python
import pandas as pd
import datetime
```

```python
file2 = pd.read_csv("monthly_feedwheat_interpd.csv")
file1 = pd.read_csv("monthly_OTH_interpd.csv")
file3 = pd.read_csv("monthly_saleyard_interpd.csv")
```

```python
file1["Date"] = pd.to_datetime(file1["Date"].astype("str"), format="%Y-%m-%d")
file2["Date"] = pd.to_datetime(file2["Date"].astype("str"), format="%Y-%m-%d")
file3["Date"] = pd.to_datetime(file3["Date"].astype("str"), format="%Y-%m-%d")
```

```python
output4 = file1.merge(file2, 
                   on='Date', 
                   how='inner',
                   sort=True)
```

```python
output4
```

```python
final_output = output4.merge(file3, 
                   on='Date', 
                   how='outer',
                   sort=True)
```

```python
final_output.to_csv("inner_join.csv")
```
